#**画像を8近傍で平均したものを出力する関数**  
**宣言**  
def myFilter_Average8( fname )  
**引数**  
fname : 平均される元の画像の名前  
**返却値**  
im_ave : 平均化された画像

```python
def myFilter_Average8( fname ):
  print( "strat:myFilter_Average8()" )
  # 画像の読み込み
  img = cv2.imread( fname )
  height , width , channel = img.shape
  print( 'width' , width )
  print( 'height' , height )
  print( 'channel' , channel )

  # 8近傍で平均化
  for ch in range( channel ):
    for y in range( 1 , height-1 , 3 ):
      for x in range( 1 , width-1 , 3 ):
        # 9マスの配列を作る(初期化値:0)
        pixels = np.zeros( 9 )
        # 中心+8近傍の画素を入れていく
        pixels[0] = img[ y , x , ch ]
        pixels[1] = img[ y-1 , x , ch ]
        pixels[2] = img[ y+1 , x , ch ]
        pixels[3] = img[ y , x-1 , ch ]
        pixels[4] = img[ y , x+1 , ch ]
        pixels[5] = img[ y-1 , x-1 , ch ]
        pixels[6] = img[ y-1 , x+1 , ch ]
        pixels[7] = img[ y+1 , x-1 , ch ]
        pixels[8] = img[ y+1 , x+1 , ch ]

        # 画素値の平均を出す
        pixels_mean = np.mean( pixels )

        # 代入
        img[ y , x , ch ] = pixels_mean
        img[ y-1 , x , ch ] = pixels_mean
        img[ y+1 , x , ch ] = pixels_mean
        img[ y , x-1 , ch ] = pixels_mean
        img[ y , x+1 , ch ] = pixels_mean
        img[ y-1 , x-1 , ch ] = pixels_mean
        img[ y-1 , x+1 , ch ] = pixels_mean
        img[ y+1 , x-1 , ch ] = pixels_mean
        img[ y+1 , x+1 , ch ] = pixels_mean

  # 返却
  print( "return:myFilter_Average8()" )
  return img

```

#**画像を32ドットのドット絵風する関数**  
**詳細**  
画像の横を32等分し、その間隔を取得。その後取得した間隔を辺の長さとした正方形で画像を分割する。そして分割された画像をそれぞれ、各画像内で最も使用されている色で染める。最後に画像を再結合したものを出力画像として返却する。  
**宣言**  
def myFilter_Dot32( fname )  
**引数**  
fname:処理を施される画像の名前  
**返却値**  
im_dot32:32ドットのドット絵風にされた画像  

```python
def myFilter_Mode32x32( fname ):
  print( "strat:myFilter_Mode32x32()" )
  # 画像を読み込む
  imgTmp = cv2.imread( fname )
  img = cv2.cvtColor( imgTmp , cv2.COLOR_BGR2RGB )
  height , width , channel = img.shape
  print( 'height' , height )
  print( 'width' , width )
  print( 'channel' , channel )

  # 表示
  plt.imshow( img )
  plt.show()

  # 分割する枚数
  split_x = 32  # 横
  split_y = 32  # 縦

  # 分割
  count_x = 0
  count_y = 0
  for i in range( split_x ):
    for j in range( split_y ):
      imgSplit = img[ count_y:count_y+int(height/split_y) , count_x:count_x+int(width/split_x) , : ]
      imgSplit_h , imgSplit_w , imgSplit_c = imgSplit.shape
#      print( "imgSplit.shape" , imgSplit.shape )
#      print( "imgSplit_h" , imgSplit_h )
#      print( "imgSplit_w" , imgSplit_w )
#      print( "imgSplit_c" , imgSplit_c )
#      print( "imgSplit[0 , 0]" , imgSplit[ 0 , 0 ] )
#      plt.imshow( imgSplit )
#      plt.show()
      for ch in range( 0 , channel , 1 ):
        # 分割した画像内の色を取得
        pixels = np.zeros( imgSplit_h*imgSplit_w )
        for y in range( 0 , imgSplit_h , 1 ):
          for x in range( 0 , imgSplit_w , 1 ):
            pixels[ y*imgSplit_h+x ] = imgSplit[ y , x , ch ]
        # 最頻値を求める
        pixels_mode_value , pixels_mode_num = stats.mode( pixels )
        # 最頻値で染める
        for y in range( 0 , imgSplit_h , 1 ):
          for x in range( 0 , imgSplit_w , 1 ):
            imgSplit[ y , x , ch ] = pixels_mode_value
      count_y = count_y + int(height/split_y)
    count_y = 0
    count_x = count_x+int(width/split_x)

  # 表示
  plt.imshow( img )
  plt.show()

```
