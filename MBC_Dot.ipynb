#**画像を8近傍で平均したものを出力する関数**  
**宣言**  
def myFilter_Average8( fname )  
**引数**  
fname : 平均される元の画像の名前  
**返却値**  
im_ave : 平均化された画像

```python
def myFilter_Average8( fname ):
  print( "strat:myFilter_Average8()" )
  # 画像の読み込み
  img = cv2.imread( fname )
  height , width , channel = img.shape
  print( 'width' , width )
  print( 'height' , height )
  print( 'channel' , channel )

  # 8近傍で平均化
  for ch in range( channel ):
    for y in range( 1 , height-1 , 3 ):
      for x in range( 1 , width-1 , 3 ):
        # 9マスの配列を作る(初期化値:0)
        pixels = np.zeros( 9 )
        # 中心+8近傍の画素を入れていく
        pixels[0] = img[ y , x , ch ]
        pixels[1] = img[ y-1 , x , ch ]
        pixels[2] = img[ y+1 , x , ch ]
        pixels[3] = img[ y , x-1 , ch ]
        pixels[4] = img[ y , x+1 , ch ]
        pixels[5] = img[ y-1 , x-1 , ch ]
        pixels[6] = img[ y-1 , x+1 , ch ]
        pixels[7] = img[ y+1 , x-1 , ch ]
        pixels[8] = img[ y+1 , x+1 , ch ]

        # 画素値の平均を出す
        pixels_mean = np.mean( pixels )

        # 代入
        img[ y , x , ch ] = pixels_mean
        img[ y-1 , x , ch ] = pixels_mean
        img[ y+1 , x , ch ] = pixels_mean
        img[ y , x-1 , ch ] = pixels_mean
        img[ y , x+1 , ch ] = pixels_mean
        img[ y-1 , x-1 , ch ] = pixels_mean
        img[ y-1 , x+1 , ch ] = pixels_mean
        img[ y+1 , x-1 , ch ] = pixels_mean
        img[ y+1 , x+1 , ch ] = pixels_mean

  # 返却
  print( "return:myFilter_Average8()" )
  return img

```

```python
def myFilter_Average24( fname ):
  print( "strat:myFilter_Average24()" )
  # 画像の読み込み
  img = cv2.imread( fname )
  height , width , channel = img.shape
  print( 'width' , width )
  print( 'height' , height )
  print( 'channel' , channel )

  # 8近傍で平均化
  for ch in range( channel ):
    for y in range( 2 , height-1 , 5 ):
      for x in range( 2 , width-1 , 5 ):
        # 25マスの配列を作る(初期化値:0)
        pixels = np.zeros( 25 )
        # 中心+24近傍の画素を入れていく
        pixels[0] = img[ y , x , ch ]
        
        pixels[1] = img[ y-1 , x , ch ]
        pixels[2] = img[ y+1 , x , ch ]
        pixels[3] = img[ y , x-1 , ch ]
        pixels[4] = img[ y , x+1 , ch ]
        pixels[5] = img[ y-1 , x-1 , ch ]
        pixels[6] = img[ y-1 , x+1 , ch ]
        pixels[7] = img[ y+1 , x-1 , ch ]
        pixels[8] = img[ y+1 , x+1 , ch ]
        
        pixels[9] = img[ y-2 , x , ch ]
        pixels[10] = img[ y+2 , x , ch ]
        pixels[11] = img[ y , x-2 , ch ]
        pixels[12] = img[ y , x+2 , ch ]
        pixels[13] = img[ y-2 , x-2 , ch ]
        pixels[14] = img[ y-2 , x+2 , ch ]
        pixels[15] = img[ y+2 , x-2 , ch ]
        pixels[16] = img[ y+2 , x+2 , ch ]

        pixels[17] = img[ y-2 , x-1 , ch ]
        pixels[18] = img[ y-2 , x+1 , ch ]
        pixels[19] = img[ y-1 , x-2 , ch ]
        pixels[20] = img[ y-1 , x+2 , ch ]
        pixels[21] = img[ y+1 , x-2 , ch ]
        pixels[22] = img[ y+1 , x+2 , ch ]
        pixels[23] = img[ y+2 , x-1 , ch ]
        pixels[24] = img[ y+2 , x+1 , ch ]

        # 画素値の平均を出す
        pixels_mean = np.mean( pixels )

        # 代入
        img[ y , x , ch ] = pixels_mean
        
        img[ y-1 , x , ch ] = pixels_mean
        img[ y+1 , x , ch ] = pixels_mean
        img[ y , x-1 , ch ] = pixels_mean
        img[ y , x+1 , ch ] = pixels_mean
        img[ y-1 , x-1 , ch ] = pixels_mean
        img[ y-1 , x+1 , ch ] = pixels_mean
        img[ y+1 , x-1 , ch ] = pixels_mean
        img[ y+1 , x+1 , ch ] = pixels_mean
        
        img[ y-2 , x , ch ] = pixels_mean
        img[ y+2 , x , ch ] = pixels_mean
        img[ y , x-2 , ch ] = pixels_mean
        img[ y , x+2 , ch ] = pixels_mean
        img[ y-2 , x-2 , ch ] = pixels_mean
        img[ y-2 , x+2 , ch ] = pixels_mean
        img[ y+2 , x-2 , ch ] = pixels_mean
        img[ y+2 , x+2 , ch ] = pixels_mean

        img[ y-2 , x-1 , ch ] = pixels_mean
        img[ y-2 , x+1 , ch ] = pixels_mean
        img[ y-1 , x-2 , ch ] = pixels_mean
        img[ y-1 , x+2 , ch ] = pixels_mean
        img[ y+1 , x-2 , ch ] = pixels_mean
        img[ y+1 , x+2 , ch ] = pixels_mean
        img[ y+2 , x-1 , ch ] = pixels_mean
        img[ y+2 , x+1 , ch ] = pixels_mean

  # 返却
  print( "return:myFilter_Average24()" )
  return img
```
